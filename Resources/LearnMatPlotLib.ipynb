```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of plots

### Line Plots

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title('simple line plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True)
```

## Bar chart

```python
months = ['January','February','March','April']
revenue = [100,200,300,500]
plt.figure(figsize=(10,6))
plt.bar(months,revenue)
plt.title('Monthly revenue chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
```

## Subplot and Legends

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8,21,25,28,38]
```

```python
plt.plot(days,original_price,'b-o',label='Original Price');
plt.plot(days,predicted_price,'r-o',label='Predicted Price');
plt.legend()
plt.title('PLot with Legends');
plt.xlabel('weekdays');
plt.ylabel('Price');
```

## Subplots

```python
#first subplot - Original price
plt.subplot(3,1,1)
plt.plot(days,original_price,'b-o',label='Original Price');
plt.title('Original Price over time');
plt.xlabel('weekdays');
plt.ylabel('Price');
plt.legend()

#second subplot - Predicted price
plt.subplot(3,1,2)
plt.plot(days,predicted_price,'r-o',label='Predicted Price');
plt.title('Predicted Price over time');
plt.xlabel('weekdays');
plt.ylabel('Price');
plt.legend()


#second subplot - The differnece between original and Predicted price
price_difference = [o-p for o, p in zip(original_price,predicted_price)]
plt.subplot(3,1,3)
plt.bar(days,price_difference,color='purple',label='Price differenece');
plt.title('differnece between original and Predicted price');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend()
plt.tight_layout()
```

## Saving the Figure 

```python
plt.plot(days,original_price,'b-o',label='Original Price');
plt.plot(days,predicted_price,'r-o',label='Predicted Price');
plt.legend()
plt.title('PLot with Legends');
plt.xlabel('weekdays');
plt.ylabel('Price');
plt.savefig('example_prediction.png',dpi=300)
```
